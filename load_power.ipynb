Calling the libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from tabulate import tabulate
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPRegressor
```

Import data from the CSV file

```python
data = pd.read_csv('https://drive.google.com/uc?id=1hrGWwY3g2Cz8OZehK9sKIkF81J7mLdoA', index_col='Dates ', parse_dates=['Dates '])
data
```

Create input/output Matrix

```python
def create_io_matrix(data, num_lags, train_ratio=0.8):
    # create input/output matrices
    x, y = [], []
    for i in range(num_lags, len(data)):
        x.append(data.iloc[i-num_lags:i, 1:4].values) 
        y.append(data.iloc[i, 1:4].values) 

    x, y = np.array(x), np.array(y)

    # Split data into training and testing sets
    train_size = int(len(x) * train_ratio)
    x_train, y_train = x[:train_size], y[:train_size]
    x_test, y_test = x[train_size:], y[train_size:]

    # Normalize input data (feature scaling)
    x_mean, x_std = x_train.mean(axis=0), x_train.std(axis=0)
    x_train = (x_train - x_mean) / x_std
    x_test = (x_test - x_mean) / x_std

    # Normalize output data
    y_mean, y_std = y_train.mean(axis=0), y_train.std(axis=0)
    y_train = (y_train - y_mean) / y_std
    y_test = (y_test - y_mean) / y_std

    stats = [x_mean, x_std, y_mean, y_std]

    return x_train, y_train, x_test, y_test, stats
```

Shape of train and test models

```python
num_lags = 3  

x_train, y_train, x_test, y_test, stats = create_io_matrix(data, num_lags)


print('x_train shape:', x_train.shape)
print('y_train shape:', y_train.shape)
print('x_test shape:', x_test.shape)
print('y_test shape:', y_test.shape)
```

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense

# Define the model architecture
model = Sequential()
model.add(LSTM(32, input_shape=(3, 2)))
model.add(Dense(1))

# Compile the model
model.compile(optimizer='adam', loss='mse')

# Train the model using the training data
model.fit(x_train, y_train, epochs=50, batch_size=32)

# Use the trained model to make predictions on the testing data
y_pred = model.predict(x_test)
```
